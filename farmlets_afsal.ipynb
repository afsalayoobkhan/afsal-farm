```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'D:\\data science exam\\afsal.csv' 
data = pd.read_csv(file_path)
```

```python
behavioral_data = data.select_dtypes(include=[float, int])
```

```python
behavioral_data['Farmlet'] = data['Farmlet']
```

```python
mean_values = behavioral_data.groupby('Farmlet').mean()
```

```python
plt.figure(figsize=(15, 8))
```

```python
melted_data = mean_values.reset_index().melt(id_vars='Farmlet', var_name='QBA Variables', value_name='Mean Score')
```

# panning to plot the mean score of each QBA variable over the three different farmlets

```python
sns.barplot(x='QBA Variables', y='Mean Score', hue='Farmlet', data=melted_data)
```

```python
sns.barplot(x='QBA Variables', y='Mean Score', hue='Farmlet', data=melted_data)

plt.title('Mean Scores of Behavioral Variables Across Different Farmlets')
plt.xlabel('Behavioral Variables')
plt.ylabel('Mean Scores')
plt.xticks(rotation=45)
plt.legend(title='Farmlet')
plt.tight_layout()
plt.show()
```
